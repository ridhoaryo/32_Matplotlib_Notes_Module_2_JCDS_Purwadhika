bentuk garis:
- ":" = titik2 di sepanjang garis
- "-" = garis biasa
- "--" = garis putus2
- "-." = garis dan titik
- "s" = kotak
- "^" = segitiga
- "*" = bintang
- "." = titik2
- "o" = titik2 besar
- "v" = segitiga ke bawah
- "v:" = segitiga ke bawah dengan titik2

```python
excl = pd.read_csv('EXCL_JK.csv', parse_dates = ['Date'], index_col='Date')
fren = pd.read_csv('FREN_JK.csv', parse_dates = ['Date'], index_col='Date')
isat = pd.read_csv('ISAT_JK.csv', parse_dates = ['Date'], index_col='Date')
tlkm = pd.read_csv('TLKM_JK.csv', parse_dates = ['Date'], index_col='Date')
```

```python
excl = excl.sort_index().resample('D').sum()
fren = fren.sort_index().resample('D').sum()
isat = isat.sort_index().resample('D').sum()
tlkm = tlkm.sort_index().resample('D').sum()
```

```python
excl.iloc[:,:5] = excl.iloc[:,:5].replace(0, np.NaN)
fren.iloc[:,:5] = fren.iloc[:,:5].replace(0, np.NaN)
isat.iloc[:,:5] = isat.iloc[:,:5].replace(0, np.NaN)
tlkm.iloc[:,:5] = tlkm.iloc[:,:5].replace(0, np.NaN)
```

```python
excl = excl.fillna(method='ffill', axis=0)
fren = fren.fillna(method='ffill', axis=0)
isat = isat.fillna(method='ffill', axis=0)
tlkm = tlkm.fillna(method='ffill', axis=0)
```

```python
excl_close = excl.Close
fren_close = fren.Close
isat_close = isat.Close
tlkm_close = tlkm.Close
```

```python
plt.figure(figsize=(15,7))
plt.plot(excl_close)
plt.plot(fren_close)
plt.plot(isat_close)
plt.plot(tlkm_close)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
